```python
import pandas as pd
import numpy as np
import os
```

```python
purchase_data = "purchase_data.csv"

purchase_data_df = pd.read_csv(purchase_data)
purchase_data_df.head()
```

```python
uniqueSN = purchase_data_df['SN'].unique()
unique_players = len(uniqueSN) 

total_players_dict = {"Total Players": [unique_players]}
total_players_df = pd.DataFrame(total_players_dict)
total_players_df
```

```python
uniqueitems = len(purchase_data_df["Item ID"].unique())
uniqueitems

numberpurchases = len(purchase_data_df["Purchase ID"].unique())
numberpurchases

totalrevenue = purchase_data_df["Price"].sum()
totalrevenue

averageprice = purchase_data_df["Price"].mean()
averageprice

summarytable = pd.DataFrame({"Unique Item": [uniqueitems],
                            "Number of Purchases": [numberpurchases],
                            "Average Price": [averageprice],
                            "Total Revenue": [totalrevenue]})
summarytable


```

```python
males_df = purchase_data_df.loc[purchase_data_df['Gender'] == "Male",:]
females_df = purchase_data_df.loc[purchase_data_df['Gender'] == "Female",:]

others_list = []
for x in range(0, len(purchase_data_df['Gender'])):
    if(purchase_data_df['Gender'][x] != "Male" and purchase_data_df['Gender'][x] != "Female"):
        others_list.append(purchase_data_df['Gender'][x])
        
others_set = set(others_list)
other_df = purchase_data_df.loc[purchase_data_df['Gender'] == others_list[0],:]

male_players = len(males_df['SN'].unique())
female_players = len(females_df['SN'].unique())
other_players = len(other_df['SN'].unique())
player_gender_counts_list = []
gender_percentages_list = []

player_gender_counts_list.append(male_players)
player_gender_counts_list.append(female_players)
player_gender_counts_list.append(other_players)
for d in range(0, len(player_gender_counts_list)):
    gender_percentages_list.append(player_gender_counts_list[d]/unique_players)

gender_demographics_dict = {
    "Gender": ["Male", "Female", "Other / Non-Disclosed"],
    "Number of Players": player_gender_counts_list,
    "Percentage": gender_percentages_list
}
gender_demographics_df = pd.DataFrame(gender_demographics_dict)
gender_demographics_df
```

```python
print(max(purchase_data_df['Age']))
print(min(purchase_data_df['Age']))
bins = [0, 10, 15, 20, 25, 30, 35, 40, 45, 50]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
arange = pd.cut(purchase_data_df['Age'], bins=bins, labels=labels, right=False)
agedf = purchase_data_df
agedf['Age Range'] = arange


age_purchase_count = []
age_total_revenue = []
age_average_purchase_price = []
for label in labels:
    agerange_df = agedf.loc[agedf['Age Range'] == label,:]
    age_purchase_count.append(len(agerange_df['Item ID']))
    age_total_revenue.append(round(sum(agerange_df['Price']), 4))
    age_average_purchase_price.append(round(sum(agerange_df['Price'])/len(agerange_df['Price']), 4))

avg_purchase_total = age_average_purchase_price




print(age_purchase_count)
print(age_total_revenue)
print(age_average_purchase_price)
print(avg_purchase_total)
```

```python
range_list = ["0-9", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40-44", "45-50"]
age_dict = {
    "Age Range": range_list,
    "Purchase Count": age_purchase_count,
    "Total Revenue": age_total_revenue,
    "Average Purchase Price": age_average_purchase_price,
    "Average Purchase Total Per Person": avg_purchase_total
}
age_results_df = pd.DataFrame(age_dict, columns=["Age Range", "Purchase Count", "Average Purchase Price", 
                                                 "Total Revenue", "Average Purchase Total Per Person"])
age_results_df
```
